```python
# provide the path of  pdf file/files.
pdfreader = PdfReader('/content/drive/MyDrive/THILINA S./Industrial Works/Bluchip Asia Technologies/NLP using NLKT and LangChain/Budget_2024.pdf')
#'/content/drive/MyDrive/THILINA S./Industrial Works/Bluchip Asia Technologies/RecoSystem/movies_metadata.csv', low_memory=False)



```

```python
from typing_extensions import Concatenate
# read text from pdf
raw_text = ''
for i, page in enumerate(pdfreader.pages):
    content = page.extract_text()
    if content:
        raw_text += content
```

```python
raw_text
```

```python
# We need to split the text using Character Text Split such that it sshould not increse token size
text_splitter = CharacterTextSplitter(
    separator = "\n",
    chunk_size = 800,
    chunk_overlap  = 200,
    length_function = len,
)
texts = text_splitter.split_text(raw_text)
```

```python
len(texts)
```

### Embeddings with huggingface api

```python
# Embeddings
embeddings = HuggingFaceEmbeddings()
```

```python
document_search = FAISS.from_texts(texts, embeddings)
```

```python
document_search
```

### Creating a Chain

```python
from langchain.chains.question_answering import load_qa_chain
from langchain import HuggingFaceHub
```

```python
from langchain.vectorstores import FAISS
document_search = FAISS.from_texts(texts, embeddings)
```

```python
chain = load_qa_chain(llm, chain_type="stuff")
```
